# Data Visualization

```python
Unique_Gender_list = student_ml_dataset_cleared['Gender'].unique()
```

```python
list_Unique_Gender_num = []
list_Unique_Gender_gender = []
for i in Unique_Gender_list:
    x = (student_ml_dataset_cleared['Gender'] == i).sum()
    print(f"{i} : {x}")
    list_Unique_Gender_num.append(x)
    list_Unique_Gender_gender.append(i)
```

```python
plt.bar(list_Unique_Gender_gender, list_Unique_Gender_num,color='black')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Male and Female')
plt.show()
```

```python
plt.pie(list_Unique_Gender_num,labels=list_Unique_Gender_gender, startangle = 90)
plt.legend()
plt.show()
```

# Correlation Between IQ and the Subjects

```python
plt.scatter(student_ml_dataset_cleared['Physics_Marks'],student_ml_dataset_cleared['IQ'], c='black', s=50, marker='.')
plt.title('IQ vs Physics_Marks')
plt.ylabel('IQ')
plt.xlabel('Physics_Marks')
plt.show()
```

```python
plt.scatter(student_ml_dataset_cleared['Math_Marks'],student_ml_dataset_cleared['IQ'], c='black', s=50, marker='.')
plt.title('IQ vs Math_Marks')
plt.ylabel('IQ')
plt.xlabel('Math_Marks')
plt.show()
```

```python
plt.scatter(student_ml_dataset_cleared['Chemistry_Marks'],student_ml_dataset_cleared['IQ'], c='black', s=50, marker='.')
plt.title('IQ vs Chemistry_Marks')
plt.ylabel('IQ')
plt.xlabel('Chemistry_Marks')
plt.show()
```
